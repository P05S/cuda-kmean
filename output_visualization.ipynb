```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./data/boxes3.csv')
df
```

# Sequential

```python
centroids = pd.read_csv('output/centroids_kmean_sequential.csv')
clusters = pd.read_csv('output/clusters_kmean_sequential.csv')
# combine df + clusters
df_plot = df.copy()
df_plot['cluster'] = clusters['cluster']

plt.figure(figsize=(8, 6))

plt.scatter(df_plot['x'], df_plot['y'],
            c=df_plot['cluster'],
            s=10,
            alpha=0.8)

plt.scatter(centroids['dim0'], centroids['dim1'],
            marker='X',
            s=200,
            linewidths=2,
            edgecolors='black')

plt.title("K-means Clustering")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True, alpha=0.2)

plt.show()
```
